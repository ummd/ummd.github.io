## (ISLR #6) [25 marks]
The regression line is given by:
\begin{array}
yy = \hat\beta_0 + \hat\beta_1 x
\end{array}

From 3.4, we know that $\hat\beta_0 = \bar y - \hat\beta_1 \bar x$. Subbing in this term:

\begin{array}
yy &= (\bar y - \hat\beta_1 \bar x) + \hat\beta_1 x \\
0 &= \bar y - y - \hat\beta_1 \bar x + \hat\beta_1 x \\
\end{array}

For the right hand side to equal zero, $(\bar{y},\bar{x})$ must be points on this regression line (i.e. $y = \bar{y}$ and $x = \bar{x}$).

```r
set.seed(1)
x=rnorm(100)
y=2*x+rnorm(100)

lm.fit = lm(y~x+0)
summary(lm.fit)
```

**[5 marks]** The coefficient is 1.9939, which is very close to the true value of 2. We can reject the null that the coefficient is zero (the p-value is small). The t-stat is 18.73 and the p-value is $<2\times 10^{-16}$

We cannot reject the null that it is 2.

```r
t= (coef(lm.fit) - 2) / summary(lm.fit)$coef[2]
t # test statistic
2*(1-pt(abs(t),100-1)) # p-value
```

The data generating process $y=2x+\epsilon$ can be re-written as: 

$$x=.5y - .5 \epsilon$$

The second regression is estimating this model. However there is an issue here, because now the error term in the model is correlated with the predictor ($y$ and  $\epsilon$ are correlated). 

Because they are positively correlated, our estiamte of the coefficient from the second regression will be too small (0.39 vs 0.5).

To see this, we can reject the null that it is 0.5. 

```r
t= (coef(lm.fit) - 0.5) / summary(lm.fit)$coef[2]
t # test statistic
2*(1-pt(abs(t),100-1)) # p-value
```

The p-value for the null of 0.5 is very small. 

Even so, the test-statistic for the null of zero is the same in both regressions.

## 11d.

**[5 marks]**
The t-statistic is given by:

$$ t = \frac{\hat\beta}{SE(\hat\beta)} $$

In this problem, there is only one predictor $x$ and no intercept, so OLS estimation is given by:

\begin{array}{cc}
\hat\beta = \frac {x^\top y} {x^\top x} \quad \text{and } SE(\hat\beta) = \sqrt{\frac {\hat\sigma} {x^\top x}}, \text{ and } \hat\sigma = \frac {e^\top e} {(n-1)}
\end{array}

### Solution using linear algebra

Because there is only one predictor, $x^\top x = \sum_{i=1}^n x_i^2$ is just a scalar (a number). This means we can write the inverse as division i.e. $(x^\top x)^{-1} = \frac{1}{x^\top x}$. This is only true for the case of one predictor.

So the t-statistic is given by:

\begin{array}
 tt &= \frac {x ^\top y} {x^\top x} \sqrt{\frac {(n-1) x^\top x} {e^\top e}} \\
 & = \frac { \sqrt{n-1} \ x ^\top y} {\sqrt{x^\top x \ e ^\top e}}
\end{array}

The term in the denominator $x^\top x \ e ^\top e$ needs to be expanded

\begin{array}
xx^\top x \ e ^\top e &= x^\top x \left( \ (y-x\hat\beta) ^\top (y-x\hat\beta) \right)\\
&= x^\top x \left( \ y^\top y -2 \hat\beta x^\top y +\hat\beta^2 x^\top x  \right)\\
&= x^\top x \left( \ y^\top y -2 \hat\beta x^\top y +\hat\beta x^\top y  \right)\\
&= x^\top x \left( \ y^\top y - \hat\beta x^\top y \right)\\
&= x^\top x \ y^\top y - \hat\beta x^\top x \ x^\top y \\
&= x^\top x \ y^\top y - x^\top y \ x^\top y \\
\end{array}

The third and fifth line come from recognizing that $x^\top x \ \hat \beta = x^\top y$.

Putting all of these terms together yields:

\begin{array}
 tt = \frac { \sqrt{n-1} \ x ^\top y} {\sqrt{x^\top x \ y^\top y - x^\top y \ x^\top y}}
\end{array}

### Solution using standard algebra

\begin{array}
tt&={\frac {\sum{x_i y_i}} {\sum{x_i^2}}}
    {\sqrt{\frac {(n-1) \sum{x_i^2}} {\sum{(y_i - x_i \beta)^2}}}}\\
    &=\frac {\sqrt{n-1} \sum{x_i y_i}}
      {\sqrt{\sum{x_i^2} \sum{(y_i - x_i \beta)^2}}} \\
&=\frac {\sqrt{n-1} \sum{x_i y_i}}
      {\sqrt{\sum{x_i^2} \sum{(y_i^2 - 2 \beta x_i y_i  + x_i^2 \beta^2)}}} \\
&=\frac {\sqrt{n-1} \sum{x_i y_i}}
      {\sqrt{\sum{x_i^2} \sum{y_i^2} - 
            \sum{x_i^2} \beta (2 \sum{x_i y_i} - \beta \sum{x_i^2})}} \\
&=\frac {\sqrt{n-1} \sum{x_i y_i}}
      {\sqrt{\sum{x_i^2} \sum{y_i^2} - 
            \sum{x_i y_i} (2 \sum{x_i y_i} - \sum{x_i y_i})}} \\
&= \frac {\sqrt{n-1} \sum{x_i y_i}} 
          {\sqrt{\sum{x_i^2} \sum{y_i^2} - (\sum{x_i y_i})^2 }}
          \end{array}
