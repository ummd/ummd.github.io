b) The irreducible error is given by the variance of the error true model, so it is constant.

$MSE_{train}$ decreases as the flexibility of the model increases because fitting methods attempt to minimize this quantitity. As the model becomes more flexible, it can get arbitrarily close to each observation, thus is able to reduce $MSE_{train}$ to zero.

$MSE_{test}$ has a u-shape, initially decreasing for inflexible models but then increasing as the model becomes more flexible. It's shape is given by the sum of two terms, $Bias^2$ and $Variance$, which tradeoff against each other as flexibility increases.

As flexibility increases,  $Bias$ reduces because a more flexible model can approach the true model, providing a better fit for the training data, thus a better prediction of the test data. However, as flexibility increases the $Variance$ also increases because the model can overfit the training data. This will lead to a worse prediction of the test data.
