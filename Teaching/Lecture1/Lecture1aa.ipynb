## The COVID Dataset (or why R is awesome)

The government of Ontario has posted COVID case data. What does the trend look like?

### Load a Dataset

```r
COVID <- read.csv("https://data.ontario.ca/dataset/f4f86e54-872d-43f8-8a86-3892fd3cb5e6/resource/ed270bb8-340b-41f9-a7c6-e8ef587e6d11/download/covidtesting.csv")
COVID #this is a dataframe
```

### Attach Variable Names to Workspace

```r
attach(COVID) #this puts the names from the data.frame into the workspae so we can call them directly

Reported.Date <- as.Date(COVID$Reported.Date, format="%Y-%m-%d") #tell R that this is a Date format, not just a string of characters, this will allow us to use some special functions on it
```

### Plot Data

```r
plot(Reported.Date,Confirmed.Positive)
```

### Fitting a Linear Model to this data

```r
lmfit <- lm(Confirmed.Positive ~ Reported.Date) # this fits the model  
summary(lmfit) #this prints a summary of the model fit

plot(Reported.Date,Confirmed.Positive) # plot the data again
abline(lmfit, col="red") # plot the fitted line from the model fit
```
