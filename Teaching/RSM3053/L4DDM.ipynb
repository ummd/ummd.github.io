```r
DDM = function(driftrate, threshold, ndt = .3, precision = .001){
  # DDM is a function for computing a single decision by using sequential evidence accumulation
  # The function returns a list of three variables:
  # 1) Decision (0 if reject, 1 if accept)
  # 2) Correct (1 if the choice matches the drift rate, 0 otherwise)
  # 3) Response Time (in milliseconds)
  # 4) The state of the evidence accumulation process at each moment
  #    until it crosses the threshold
  # DDM requires specification of:
  #  Standard DDM parameters: driftrate, threshold, and ndt (in secs)
  #  precision: time-step of the simulation (default = 1ms), used to scale other parameters

  # Ex 1.1. (insert response here)
  Evidence = threshold/2

  t = 1 # initialize the time counter at time t = 1
  Decision = NA # initialize a decision variable

  # Ex 1.2: (insert response here)
  while(is.na(Decision)){

    # Ex. 1.3a: (insert response here)
    drift = driftrate*precision + rnorm(1, mean = 0, sd = 1 * sqrt(precision))
    # Ex. 1.3b: (insert respose here)
    Evidence[t + 1] = Evidence[t] + drift

    # Ex. 1.4: (insert response here)
    if(Evidence[t + 1] > threshold | Evidence[t+1] < 0 ){
      Decision = Evidence[t + 1] > 0
    }
    t = t + 1 # increment the time step
  } # end of while loop

  Correct = Decision == (driftrate > 0) # determine accuracy

  # Ex. 1.5: (insert response here)
  RT = length(Evidence)*precision + ndt

  # adjust evidence trace to include NDT
  Evidence = c(rep(Evidence[1],ceiling(ndt/precision)), Evidence)

  # create a list containing the data on decision, RT, and the path traced by the accumulated evidence
  ddm = list(Decision = Decision, Correct = Correct, RT = RT, driftTrace = Evidence)
}
```
